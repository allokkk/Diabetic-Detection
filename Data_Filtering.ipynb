```
listimages=[]
```

```
for i in items:
  j=i.split(".")
  print(j[0])
  listimages.append(j[0])
```

```
x_train=pd.read_csv("/content/drive/My Drive/trainLabels.csv")
print(x_train.shape)
print(len(listimages))
```

```

x_train.head(n=12)
```

```
listimages2=[]
listimages3=[]
for i in listimages:
  for j in x_train["image"]:
    
    if i==j:
      listimages2.append(j)
      y=x_train.loc[x_train['image'] == j, 'level'].iloc[0]
      listimages3.append(y)
    
```
